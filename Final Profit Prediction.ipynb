```python
X= dataset.iloc[:,:-1].values
y=dataset.iloc[:,3].values
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labelencoder = LabelEncoder()
X[:,2] = labelencoder.fit_transform(X[:,2])
X1 = pd.DataFrame(X)
X1.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,train_size=0.7,random_state=0)
x_train
```

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
model.fit(x_train,y_train)
print('model has been trained succesfully')
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
testing_data_model_score= model.score(x_test,y_test)
print("model score on testing data",testing_data_model_score)
```

```python
df = pd.DataFrame(data={'predicted value':y_pred.flatten(),'Actual value':y_test.flatten()})
df
```

```python
from sklearn.metrics import r2_score
r2Score= r2_score(y_pred,y_test)
print("R2 Score of MODEL:-",r2Score*100)
```

```python
from sklearn.metrics import mean_squared_error
Mean_Squared_Error = mean_squared_error(y_pred,y_test)
print("MEAN SQUARED ERROR :-", Mean_Squared_Error*100)
```

```python
import numpy as np
rmse = np.sqrt(mean_squared_error(y_pred,y_test))
print("ROOT MEAN SQUARED ERROR :-", rmse * 100)         
```

```python
from sklearn.metrics import mean_absolute_error
Mean_Absolute_Error = mean_absolute_error(y_pred,y_test)
print("MEAN ABSOLUTE ERROR :-",Mean_Absolute_Error)
```
